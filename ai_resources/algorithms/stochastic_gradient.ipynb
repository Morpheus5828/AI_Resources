```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
# Stochastic gradient
```

```python
def stochastic_gradient(
    X: np.ndarray,
    f_prime,
    n: int,
    learning_rate: float = 1e-3,
    max_iterations: int=1000
):
    x = np.zeros(X.shape[1])
    for _ in range(max_iterations):
        i = random.randint(0, n-1)
        x = x - learning_rate * f_prime(X[i])
    return x
```

```python
def f_prime(x):
    return 2*x
```

```python
n = 100
X = np.arange(-n, n).reshape(-1, 1)
X.shape
```

```python
min = stochastic_gradient(
    X=X,
    f_prime=f_prime,
    n=n,
    learning_rate=1e-5
)
```

```python
plt.plot(X, X**2)
plt.scatter(min, min**2)
```
