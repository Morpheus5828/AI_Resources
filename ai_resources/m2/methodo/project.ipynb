```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
path = "Converted_Data.csv"
data = pd.read_csv(path, sep=",")

# Convert all relevant columns to numeric, replacing commas and handling non-numeric values
data.replace(',', '.', regex=True, inplace=True)
data_numeric = data.apply(pd.to_numeric, errors='coerce')
```

```python
columns_to_drop = ['Lang code', 'Lang name', 'Mono+c_stdev', 'Multi+c_stdev',
                   'Multi+c_W22_stdev', 'Multi−c_W22_stdev', 'ZS+c_stdev',
                   'ZS−c_stdev', 'Mono−c_stdev', 'Multi−c_stdev', 'Multi+c_ID_stdev',
                   'Multi−c_ID_stdev', 'ZS+c_W22_stdev', 'ZS−c_W22_stdev']

data_numeric = data.drop(columns=columns_to_drop)

''''''
```

```python
data_numeric.info()
data_numeric["Multi+c_W22_acc"] = pd.to_numeric(data_numeric["Multi+c_W22_acc"])
data_numeric["Multi−c_W22_acc"] = pd.to_numeric(data_numeric["Multi−c_W22_acc"])
data_numeric["Multi+c_ID_acc"] = pd.to_numeric(data_numeric["Multi+c_ID_acc"])
```

```python
import numpy as np
for m in data_numeric.columns:
    plt.figure(figsize=(20, 10))
    plt.xticks(range(data_numeric[m].to_numpy().shape[0]), data["Lang name"].to_list(), rotation=45, ha='right')
    plt.plot(range(data_numeric[m].to_numpy().shape[0]), data_numeric[m].to_numpy())
    plt.title(m)
    print(m, np.std(data_numeric[m].to_numpy()))
    plt.ylabel("Accuracy")
    plt.plot(range(data_numeric[m].to_numpy().shape[0]), [np.mean(data_numeric[m].to_numpy()) for _ in range(data_numeric[m].to_numpy().shape[0])], c="red", label=f"Mean: {np.mean(data_numeric[m].to_numpy()):.2f}")
    plt.legend()
    plt.savefig(m)
```

```python
col = ["Mono+c_acc", "Multi+c_W22_acc", "Multi−c_W22_acc", "Multi+c_ID_acc", "Multi−c_ID_acc", "ZS+c_acc", "ZS−c_acc", "ZS+c_W22_acc", "ZS−c_W22_acc"]
```

```python
plt.figure(figsize=(20, 10))
plt.subplots_adjust(bottom=0.3)
for m in col:
    plt.xticks(range(data_numeric[m].to_numpy().shape[0]), data["Lang name"].to_list(), rotation=45, ha='right')
    plt.plot(range(data_numeric[m].to_numpy().shape[0]), data_numeric[m].to_numpy(), label=m)
plt.legend()
plt.ylabel("Accuracy")
plt.savefig("acc.png")
```

```python
plt.figure(figsize=(20, 10))
plt.xticks(range(data_numeric["Mono+c_acc"].to_numpy().shape[0]), data["Lang name"].to_list(), rotation=45, ha='right')
plt.plot(range(data_numeric["Mono+c_acc"].to_numpy().shape[0]), data_numeric["Mono+c_acc"].to_numpy(), label="Mono+c_acc", c="blue")
plt.plot(range(data_numeric["Mono−c_acc"].to_numpy().shape[0]), data_numeric["Mono−c_acc"].to_numpy(), label="Mono-c_acc", c="red")
plt.ylabel("Accuracy")
plt.legend()
plt.savefig("c1.png")
```

```python
plt.figure(figsize=(20, 10))
plt.xticks(range(data_numeric["Multi+c_acc"].to_numpy().shape[0]), data["Lang name"].to_list(), rotation=45, ha='right')
plt.plot(range(data_numeric["Multi+c_acc"].to_numpy().shape[0]), data_numeric["Multi+c_acc"].to_numpy(), label="Multi+c_acc", c="blue")
plt.plot(range(data_numeric["Multi−c_acc"].to_numpy().shape[0]), data_numeric["Multi−c_acc"].to_numpy(), label="Multi-c_acc", c="red")
plt.ylabel("Accuracy")
plt.legend()
plt.savefig("c2.png")
```

```python
plt.figure(figsize=(20, 10))
plt.xticks(range(data_numeric["ZS+c_acc"].to_numpy().shape[0]), data["Lang name"].to_list(), rotation=45, ha='right')
plt.plot(range(data_numeric["ZS+c_acc"].to_numpy().shape[0]), data_numeric["ZS+c_acc"].to_numpy(), label="ZS+c_acc", c="blue")
plt.plot(range(data_numeric["ZS−c_acc"].to_numpy().shape[0]), data_numeric["ZS−c_acc"].to_numpy(), label="ZS-c_acc", c="red")
plt.ylabel("Accuracy")
plt.legend()
plt.savefig("c3.png")
```
